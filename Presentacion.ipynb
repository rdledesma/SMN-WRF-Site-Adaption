# Fit a Multilayer Percetron for improve estimation from WRF-SMN

En este script se detallan los pasos a seguir para entrenear un modelo de red neuronal para mejorar la estimación de GHI del modelo WRF utilizando medidas en tierra. 

### Paso 1. Descarga de las estimaciones del modelo WRF-SMN

En la documentación oficial https://odp-aws-smn.github.io/documentation_wrf_det/Acceso_a_los_datos/ se ofrecen diferentes opciones para la descarga de los archivos netCDF generados por el modelo WRF-SMN

Este tutorial está completamente realizado en el lenguaje python instalados en una pc con la distribución linux mint como sistema operativo.  
Las dependencias necesarias son:
- python 3.15.5
- pandas
- matplotlib
- numpy
- xarray 

Adicionalmente, sugerimos una alternativa que permite descargar de manera eficiente grande volúmenes de datos del modelo WRF-SMN.

```
aws s3 sync --no-sign-request "s3://smn-ar-wrf/DATA/WRF/DET/2024" images/2024/ \
  --exclude "*10M*.nc" \
  --exclude "*24H*.nc" \
  --only-show-errors 
```

El código anterior permite descargar $todos$ los datos almacenados para el año $2024$ y guardarlos en el directorio $images/2024$
